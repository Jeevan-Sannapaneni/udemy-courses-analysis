```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x="subject",order=df['subject'].value_counts().index,palette="viridis")
plt.title("Number of Courses per Subject")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['price'],bins=30,kde=True,color="blue")
plt.title("Distribution of Course Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x="price",y="num_subscribers",alpha=0.6)
plt.title("Price vs Number of Subscribers")
plt.xlabel("Price")
plt.ylabel("Subscribers")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x="level",y="content_duration",palette="Set2")
plt.title("Content Duration by Course Level")
plt.xticks(rotation=45)
plt.show()
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df['year'] = df['published_timestamp'].dt.year
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="year",palette="coolwarm")
plt.title("Number of Courses Published per Year")
plt.xticks(rotation=45)
plt.show()
```
